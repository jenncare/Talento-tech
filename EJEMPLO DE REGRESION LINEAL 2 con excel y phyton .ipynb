```python
import pandas as pd


# dataset sera esta
dato1 = [1,2,2,3,4,4,5,6]
dato2 = [2,3,4,4,4,6,5,7]

m= 0.849
b= 1.504

diccionario = {"x": dato1, "y": dato2}

df= pd.DataFrame(diccionario)

df
```

```python
# Para realizar tabla de excel, para calcular ypred

df ["ypred"] = m * df.x + b

df

# Para realizar tabla de excel, para calcular y-ypred

df ["y-ypred"] = df.y - df.ypred

df

# Para realizar tabla de excel, para calcular ypred-y

df ["ypred-y"] = df.ypred - df.y

# Para realizar tabla de excel, para calcular ypred-y

df ["error2"] = df["ypred-y"] **2

df
```
